### The mmMapPlot class plots x/y data from a map.

### Load a map

```python
from PyMapManager.mmMap import mmMap

# load a map
filePath = '/Users/cudmore/Desktop/data/rr30a/rr30a.txt'
m = mmMap(filePath=filePath)
```

### Import required libraries

```python
%matplotlib notebook
import matplotlib.pyplot as plt

from PyMapManager.mmMapPlot2 import mmMapPlot2
from PyMapManager.mmUtil import newplotdict
```

### Plot a connonical dendrogram of all spines and their position along the segment (dendrite) tracing.

```python
plotDict = newplotdict()
plotDict['segmentid'] = [0]

mp = mmMapPlot2(m)
fig = plt.figure()
mp.plotMap0(fig, plotDict)
```

### Plot z versus imaging session days

```python
plotDict = newplotdict()
plotDict['xstat'] = 'days'
plotDict['ystat'] = 'z'
plotDict['segmentid'] = []
#plotDict['stacklist'] = [0,1,2]
mp = mmMapPlot2(m)
fig = plt.figure()
mp.plotMap(fig, plotDict)
```

### Plot spine intensity versus segment intensity

```python
plotDict = newplotdict()
plotDict['xstat'] = 'ubssSum_int1'
plotDict['ystat'] = 'ubsdSum_int1'
plotDict['segmentid'] = []
plotDict['showlines'] = False
plotDict['showdynamics'] = False

mp = mmMapPlot2(m)
fig = plt.figure()
mp.plotMap(fig, plotDict)
```
