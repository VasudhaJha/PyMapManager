```python
# copy and paste this into ipython command line
#%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
from numpy.random import rand

m = 5
n = 3
x = rand(m,n)
y = rand(m,n)
c = rand(m,n)
s = rand(m,n)

for i in range(m):
    for j in range(n):
        x[i][j] = i*n + j
        y[i][j] = i

x[2,:] = 'nan'
y[2,:] = 'nan'

xLine = x.transpose()
yLine = y.transpose()


def onpick3(event):
    ind = event.ind
    print('** onpick3 scatter:', ind, np.take(x, ind), np.take(y, ind))
    print ind

fig, ax = plt.subplots()
col = ax.scatter(x, y, 100*s, c, picker=True)

lines = ax.plot(xLine, yLine, '-')

fig.canvas.mpl_connect('pick_event', onpick3)

plt.show()
```
