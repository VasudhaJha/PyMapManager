### Time-series annotations and images can be loaded directly from a remote server.

Here we will load and plot a map from server at http://cudmore.duckdns.org

```python
from pymapmanager import mmio
print(mmio.server)
print(mmio.user)
```

### Load map from remote server

```python
from pymapmanager import mmMap
urlmap = 'rr30a'
myMap = mmMap(urlmap=urlmap)
```

### Plot

```python
from pymapmanager.mmMapPlot2 import mmMapPlot2
from pymapmanager.mmUtil import newplotdict
import matplotlib.pyplot as plt

mp = mmMapPlot2(myMap) # initialize map plot with myMap
plotDict = newplotdict() # define a plot dictionary to say what to plot
plotDict['segmentid'] = 2 # only map segment 1
mp.plotMap0(plt.figure(), plotDict)
```

### Now load locally and compare

```python
# load locally
localMap = mmMap('exampleMaps/rr30a')
localMapPlot = mmMapPlot2(localMap)

# plot local
localMapPlot.plotMap0(plt.figure(1), plotDict)


# plot remote
mp.plotMap0(plt.figure(2), plotDict)
```
