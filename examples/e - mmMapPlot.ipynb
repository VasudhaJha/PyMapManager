### The mmMapPlot class plots x/y data from a map.

### Load a map

```python
from pymapmanager.mmMap import mmMap

filePath = 'exampleMaps/rr30a/rr30a.txt'
m = mmMap(filePath=filePath)
```

### Import required libraries

```python
%matplotlib notebook
import matplotlib.pyplot as plt

from pymapmanager.mmMapPlot2 import mmMapPlot2
from pymapmanager.mmUtil import newplotdict
```

### Plot a connonical dendrogram of all spines and their position along the segment (dendrite) tracing.

```python
#mmMapPlot2.plotMap0() plots a cannonical spine map

plotDict = newplotdict()
plotDict['segmentid'] = 0

mp = mmMapPlot2(m)
fig = plt.figure()
mp.plotMap0(fig, plotDict)
```

### Plot annotation z-depth (slice) versus imaging session days

```python
plotDict = newplotdict()
plotDict['xstat'] = 'days'
plotDict['ystat'] = 'z'
plotDict['segmentid'] = None # None for all
#plotDict['stacklist'] = [0,1,2]
mp = mmMapPlot2(m)
fig = plt.figure()
mp.plotMap(fig, plotDict)
```

### Plot spine intensity versus segment intensity

```python
plotDict = newplotdict()
plotDict['xstat'] = 'ubssSum_int1' # backgorund subtracted spine intensity
plotDict['ystat'] = 'ubsdSum_int1' # backgorund subtracted segment intensity
plotDict['segmentid'] = None
plotDict['showlines'] = False
plotDict['showdynamics'] = False

mp = mmMapPlot2(m)
fig = plt.figure()
mp.plotMap(fig, plotDict)
```
